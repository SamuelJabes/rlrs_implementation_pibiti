```python
chat = ChatOpenAI(temperature=0.001, api_key=OPENAI_API_KEY, model="gpt-3.5-turbo-0125")
def get_movie_summary(movie_title):
    base_rule = 'Você é um redator que escreve resumos sobre obras cinematográficas (filmes).'
    rule_1 = f'Você irá escrever um resumo sobre o filme: {movie_title}'
    rule_2 = 'Comece com uma breve introdução ao filme, incluindo o título, o ano de lançamento (se já estiver definido) e os gêneros principais.'
    rule_3 = 'Descreva a premissa básica do filme, incluindo o cenário e os personagens principais.'
    rule_4 = 'Destaque os principais membros do elenco e seus papéis no filme.'
    rule_5 = 'Mencione o diretor do filme e, se relevante, detalhes sobre a produção, como produtores ou roteiristas renomados.'
    rule_6 = 'Se o filme apresentar alguma inovação técnica, conceitual ou narrativa, destaque isso no resumo.'
    rule_7 = 'Se houver planos para transformar o filme em uma franquia ou criar sequências, mencione isso e forneça informações relevantes sobre os planos futuros.'
    rule_8 = 'Se o filme estiver gerando expectativas significativas entre o público ou a crítica, mencione isso no resumo.'
    rule_9 = 'Se o filme estiver situado em um contexto histórico ou cultural específico, ou se abordar temas relevantes para a sociedade atual, mencione isso no resumo.'
    rule_10 = 'Se o filme já foi exibido em prévias ou festivais de cinema e recebeu feedbacks significativos, mencione isso no resumo.'
    rule_11 = 'Termine o resumo de forma concisa, reiterando os pontos principais e deixando o leitor com uma ideia clara do que esperar do filme.'
    rule_12 = 'O resumo deve ter no máximo 250 palavras'
    rule_13 = f"Aqui está um exemplo de resumo para o filme Sabrina: Sabrina (mesmo título no original em inglês) é um filme estadunidense e alemão de 1995, do gênero comédia romântica e comédia dramática, escrito por Barbara Benedek e David Rayfiel. Foi dirigido por Sydney Pollack, e estrelado por Harrison Ford como Linus Larrabee, Julia Ormond como Sabrina e Greg Kinnear (em seu primeiro papel no cinema) como David Larrabee. Também conta com Angie Dickinson, Richard Crenna, Nancy Marchand, Lauren Holly, John Wood, Dana Ivey, e a atriz francesa Fanny Ardant. A trilha sonora foi composta por John Williams e inclui uma música interpretada por Sting, Moonlight; ambos foram indicados para o Oscar 1996.\nÉ um remake do filme Sabrina de 1954, co-escrito e dirigido por Billy Wilder, estrelado por Humphrey Bogart, Audrey Hepburn e William Holden, que por sua vez foi baseado em uma peça intitulada Sabrina Fair. Nesta nova versão houve modificações em relação ao original, foi eliminado o pai dos Larrabee, aumentando o poder e a responsabilidade de Linus. No filme de 1954, David já casou três vezes; no de 1995, nenhuma, o que reforça seu perfil de playboy irresponsável. Na primeira versão, Sabrina foi a Paris estudar culinária; na segunda, foi estudar produção de moda e acabou virando fotógrafa.O local usado para retratar a mansão da família Larrabee foi a propriedade 'Salutation', localizada em Long Island, em Glen Cove, Nova York. Esta casa foi construída por volta de 1929 por Junius Spencer Morgan III, que era diretor da Morgan Guaranty Trust Company. Seu pai era J. P. Morgan, Jr., que era um banqueiro e filho do J. P. Morgan. A propriedade não é mais propriedade da família Morgan, mas ainda está em mãos privadas e usada como residência. O filme fez uso extensivo dos interiores desta mansão durante as filmagens.Sabrina foi lançada em 15 de dezembro de 1995 pela Paramount Pictures. O filme foi uma decepção de bilheteria, mas recebeu críticas positivas dos críticos. O filme teve comparações inevitáveis \u200b\u200bcom a versão original, com seu trio de estrelas, Audrey Hepburn, Humphrey Bogart e William Holden. No entanto, os críticos deram ao filme críticas positivas, com uma nova pontuação de Rotten Tomatoes de 65% com base em 48 avaliações. Se baseie no resumo."
    messages = [
        SystemMessage(content=base_rule),
        SystemMessage(content=rule_1),
        SystemMessage(content=rule_2),
        SystemMessage(content=rule_3),
        SystemMessage(content=rule_4),
        SystemMessage(content=rule_5),
        SystemMessage(content=rule_6),
        SystemMessage(content=rule_7),
        SystemMessage(content=rule_8),
        SystemMessage(content=rule_9),
        SystemMessage(content=rule_10),
        SystemMessage(content=rule_11),
        SystemMessage(content=rule_12),
        SystemMessage(content=rule_13),
        SystemMessage(content=f'Aqui está o resumo do filme {movie_title}: '),
    ]
    resumo = chat(messages).content
    return resumo
```

## Embeddings

Aqui, nós usaremos um modelo de embedding da `OpenAI`.

```python
item_num = 3883
```

```python
movie_descriptions = [] # list of movie descriptions
for i in range(0, item_num): # for movie user
    movie = movies.iloc[i, :]
    description = f"Esse é o filme de ID {movie['MovieID']}. O título deste filme é {movie['Title']}. Os gêneros dos quais esse filme se encaixa são {movie['Genres']} e uma breve descrição deste filme é a seguinte: {movie['Description']}."
    movie_descriptions.append(description)
```

```python
movie_descriptions[:5]
```

```python
model_embedding = OpenAIEmbeddings(api_key=OPENAI_API_KEY, model="text-embedding-3-large", dimensions=256)
```

```python
embeddings_descriptions = model_embedding.embed_documents(movie_descriptions)
```

```python
embeddings_descriptions = np.array(embeddings_descriptions)
embeddings_descriptions
```

```python
embeddings_descriptions.shape
```

```python
embeddings_descriptions_df = pd.DataFrame(embeddings_descriptions)
embeddings_descriptions_df.head(5)
```

```python
df_movies_with_descriptions_and_embeddings = pd.concat([movies, embeddings_descriptions_df], axis=1)
df_movies_with_descriptions_and_embeddings.head(5)
```

```python
df_movies_with_descriptions_and_embeddings.to_csv(os.path.join(DATA_DIR, "movies_with_descriptions_and_embeddings_256.csv"))
```
